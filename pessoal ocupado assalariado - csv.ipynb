```python
#Demografia das Empresas e Estatisficas de Empreendedorismo 
#2722 - Número de unidades locais e pessoal ocupado assalariado, segundo as seções da classificação de atividades

import csv
import requests

url_template = "https://servicodados.ibge.gov.br/api/v3/agregados/2722/periodos/{year}/variaveis/484?localidades=N1[all]&classificacao=12762[all]"

headers = ["ano", "variavel", "unidade", "classificacoes-nome", "classificacoes-categoria", "localidade-nome", "serie-valor"]
rows = []

for year in range(2008, 2021):
    url = url_template.format(year=year)
    response = requests.get(url)
    data = response.json()
    
    for item in data:
        variavel = item["variavel"]
        unidade = item["unidade"]
        
        for resultado in item["resultados"]:
            classificacoes = resultado["classificacoes"]
            categoria = classificacoes[0]["categoria"]
            categoria_nome = list(categoria.keys())[0]
            categoria_valor = categoria[categoria_nome]
            
            localidade = resultado["series"][0]["localidade"]["nome"]
            serie_valor = resultado["series"][0]["serie"][str(year)]
            
            row = [year, variavel, unidade, classificacoes[0]["nome"], categoria_valor, localidade, serie_valor]
            rows.append(row)

# Escrever os dados no arquivo CSV
with open("pessoalocupadoassalariado.csv", "w", newline="",encoding="utf-8-sig") as csvfile:
    writer = csv.writer(csvfile)
    writer.writerow(headers)
    writer.writerows(rows)
    
print("CSV gerado com sucesso!")
```
