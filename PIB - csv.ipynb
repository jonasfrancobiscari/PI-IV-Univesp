```python
#Contas Nacionais Anuais
#Produto Interno Bruto, Produto Interno Bruto per capita, População residente e Deflator (6784)
#Todos os periodos

import csv
import json
import requests

url = "https://servicodados.ibge.gov.br/api/v3/agregados/6784/periodos/2012|2013|2014|2015|2016|2017|2018|2019|2020/variaveis/9808|9809|9810|9811|93|9812|9813|9814?localidades=N1[all]"

response = requests.get(url)
data = response.json()

filename = "PIB.csv"

with open(filename, mode='w', encoding='utf-8-sig', newline='') as file:
    writer = csv.writer(file)
    
    # Write CSV header
    writer.writerow(["Variavel", "Unidade", "Ano", "Valor"])
    
    # Iterate through the data
    for item in data:
        variavel = item["variavel"]
        unidade = item["unidade"]
        
        resultados = item["resultados"]
        for resultado in resultados:
            series = resultado["series"]
            for serie in series:
                localidade = serie["localidade"]["nome"]
                serie_data = serie["serie"]
                
                for ano, valor in serie_data.items():
                    writer.writerow([variavel, unidade, ano, valor])
                    
print("CSV file generated successfully.")
```
